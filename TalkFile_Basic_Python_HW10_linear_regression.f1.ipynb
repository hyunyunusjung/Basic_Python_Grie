```python
import pandas as pd
import numpy as np
data = pd.read_csv('weatherHistory.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
y = data['Temperature (C)']
```

```python
y
```

```python
features = data[['Apparent Temperature (C)', 'Humidity', 'Pressure (millibars)', 'Visibility (km)', 'Wind Speed (km/h)']]
```

```python
for i in features:
    x = data[[i]]
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    print(f'{i}: R2 = {model.predict(x)}')
```

```python
print ( model.summary())
```
