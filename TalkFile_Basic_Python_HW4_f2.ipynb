**문제1) 다음 코드의 실행 결과를 맞춰보세요.**

```python
number = 196

rem = rev = 0
while number >= 1:
    rem = number % 10
    rev = rev * 10 + rem
    number = number // 10

print(rev)

# 691
```

**문제5) input()으로 사용자의 출생 연도를 입력받은 후, 다음 표의 어느 세대에 속하는지 출력하세요.**



```python
n = int(input("출생 연도: "))
if (n<= 1924):
    print("가장 위대한 세대(The Greatest Generation")
elif (n>= 1925 and n<=1945):
    print("침묵 세대 (The Silent Generation)")
elif (n>= 1946 and n<=1964):
    print("베이비붐 세대 (Baby Boomer)")
elif (n>= 1965 and n<=1980):
    print("X세대 (Generation X)")
elif (n>= 1981 and n<=1996):
    print("멜리니얼 (Millennial)")
else:
    print("Z 세대 (Generation Z)")
```
