# 01 Ramom Forest & KNN
random forest의 hyperparameter를 설정해봅시다. 지오가 원하는 hyperparameter를 하나 골라봅시다. hyperparameter의 값은 5종류로 설정해봅시다. 그럼 서로 다른 hyperparameter를 가진 random forest 5개가 만들어지겠죠?
knn의 hyperparameter k 값을 각각 다르게 5개의 서로 다른 knn을 만들어봅시다.
이 10개의 모델 중 titanic dataset의 생존 예측 성능이 가장 좋은 건 누구인가요? (전처리 하는 거 잊지 마세요!)

참조 : https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('titanic.csv')
df = data[['Pclass', 'Sex' , 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]

data['Age'] = data['Age'].fillna(data['Age'].mean())
mapping = {'male':0, 'female':1}
df['Sex'] = df['Sex'].map(mapping)

X = df[['Pclass', 'Sex', 'Age','Fare']]
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state= None)
score = 0
model = ""

for n in [50, 100, 150, 200, 250]:
    model = RandomForestClassifier(n_estimators = n)
    model.fit(X_train, y_train)
    s = model.score(X_test, y_test)
    if s > score:
        score = s
        model = f"Random Forest (n_estimators={n})"

print(f" 최고 모델: {model}")
print(f" 정확도 : {score}")
```

## 02 SVM
scikit-learn 와인 데이터셋을 각기 다른 hyperparameter (kernel, C, gamma) 조합의 SVM으로 분류해보고, 가장 성능이 좋았던 hyperparameter 조합을 알려주세요.

https://noanomal.tistory.com/81?
- 위 링크는 와인 데이터셋을 사용하는 방법을 간단하게 정리한 블로그입니다.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
import numpy as np

wine = load_wine()

df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['class'] = wine.target

X = wine.data
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state= None)

svm = SVC()
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
y_pred == y_test
print(np.mean(y_pred == y_test))
```

```python
#hyperparameter setting 1 : kernel의 종류 차이에 따른 분류 성능

svm_linear = SVC(kernel = 'linear')
svm_poly = SVC(kernel = 'poly')
svm_rbf = SVC(kernel = 'poly')

svm_linear.fit(X_train, y_train)
svm_poly.fit(X_train, y_train)
svm_rbf.fit(X_train, y_train)
np.mean(svm_linear.predict(X_test) == y_test)

print(f'prediction accuracy of linear kernel is {np.mean(svm_linear.predict(X_test) == y_test)}')
print(f'prediction accuracy of poly kernel is {np.mean(svm_poly.predict(X_test) == y_test)}')
print(f'prediction accuracy of rbf kernel is {np.mean(svm_rbf.predict(X_test) == y_test)}')
```

```python
#hyperparameter setting 2 : C의 차이에 따른 분류 성능

svm_c1 = SVC(C=1)
svm_c5 = SVC(C=5)
svm_c10 = SVC(C=10)

svm_c1.fit(X_train, y_train)
svm_c5.fit(X_train, y_train)
svm_c10.fit(X_train, y_train)

print(f'prediction accuracy of C=1 is {np.mean(svm_c1.predict(X_test) == y_test)}')
print(f'prediction accuracy of C=5 is {np.mean(svm_c5.predict(X_test) == y_test)}')
print(f'prediction accuracy of C=10 is {np.mean(svm_c10.predict(X_test) == y_test)}')
```

```python
# hyperparameter setting 3 : gamma의 차이에 따른 분류 성능

svm_gamma01 = SVC(gamma=0.1)
svm_gamma1 = SVC(gamma=1)
svm_gamma10 = SVC(gamma=10)

svm_gamma01.fit(X_train, y_train)
svm_gamma1.fit(X_train, y_train)
svm_gamma10.fit(X_train, y_train)

print(f'prediction accuracy of gamma=0.1 is {np.mean(svm_gamma01.predict(X_test) == y_test)}')
print(f'prediction accuracy of gamma=1 is {np.mean(svm_gamma1.predict(X_test) == y_test)}')
print(f'prediction accuracy of gamma=10 is {np.mean(svm_gamma10.predict(X_test) == y_test)}')
```
