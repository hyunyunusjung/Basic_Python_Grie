# 02 pandas 추가 문제
다음은 가상의 쇼핑몰 데이터입니다.

```python
data = {
    "customer_id": [1, 2, 3, 1, 2, 3],
    "product": ["A", "B", "A", "C", "A", "B"],
    "price": [10000, 20000, 10000, 15000, 10000, 20000],
    "quantity": [1, 2, 1, 3, 2, 1]
}
df = pd.DataFrame(data)
```

```python
df['total_price'] =  df["price"] * df['quantity']
```

```python
total = 0
for i in range(len(df)):
    total += df['total_price'][i]
print(total)
```

```python
product_list = []

for i in range (len(df['product'])):
    value = df['product'].iloc[i]
    if (value != 'A' and value != 'B' and value != 'C'):
        product_list.append(int(value))

count_a = 0
count_b = 0
count_c = 0

for value in product_list:    
    if ( value % 20000 == 0):
        count_b += 1
    elif ( value % 15000 == 0):
        count_c += 1
    elif (value % 10000 == 0):
        count_a += 1
    
if (count_a >= count_b) and (count_a >= count_c):
    print("A")
elif (count_b >= count_a) and (count_b >= count_c):
    print("B")
else:
    print("C")
```
