```python
from sklearn.datasets import load_iris
iris = load_iris()
feature = iris['data']
labels = iris['target']
names = iris['target_names']
import pandas as pd
feature_names = iris['feature_names']
df = pd.DataFrame(feature, columns=feature_names)
df['target'] = labels
i = 0
targetnames = []
for i in range(i,len(labels)):
    if( labels[i] == 0 ):
        targetnames.append(names[0])
    elif ( labels[i] == 1):
        targetnames.append(names[1])
    else:
        targetnames.append(names[2])
df['target names'] = targetnames
df

#targetnames = ['']*150
#for i in range(i,len(labels)):
#    if( labels[i] == 0 ):
#       targetnames[i] = (names[0])
#    elif ( labels[i] == 1):
#        targetnames[i] = (names[1])
#    else:
#        targetnames[i] = (names[2])
#df['target names'] = targetnames
#df
```
