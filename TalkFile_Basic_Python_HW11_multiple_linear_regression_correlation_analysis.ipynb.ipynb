### 문제 1
https://www.kaggle.com/datasets/budincsevity/szeged-weather 위의 데이터셋을 사용해서 Multiple Linear Regression을 이용해 예측을 해봅시다. train-test split을 하고, feature는 마음에 드는 걸 골라주세요. test dataset으로 성능을 보고해주세요.

```python
import pandas as pd
import numpy as np
data = pd.read_csv('weatherHistory.csv')

from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

from sklearn.model_selection import train_test_split
```

```python
#data.columns
df = data[['Temperature (C)', 'Apparent Temperature (C)', 'Visibility (km)', 'Humidity']]
y = df[['Temperature (C)']]
x = df[['Apparent Temperature (C)', 'Visibility (km)', 'Humidity']]
model = sm.OLS(y, np.asarray(x)).fit()
#print(model.summary())
model = LinearRegression()
train_x, test_x, train_y, test_y = train_test_split(x.values, y, random_state=15)
model.fit(train_x, train_y)
score = model.score(test_x, test_y)
print(f'Test R2: {score}')
```

### 문제 2
titanic data를 활용해 생존 여부와 다른 feature들의 상관계수를 scipy, numpy, pandas 세 가지 방법으로 구해봅시다.

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
data = pd.read_csv('titanic.csv')
df = data[['Survived','Fare']]
print(data.columns)
```

```python
pearsonr(df['Survived'], df['Fare'])
```

```python
np.corrcoef(df['Survived'], df['Fare'])[0][1]
```

```python
df_sepal = df.iloc[:, 0:2]
df_sepal.corr(method='pearson')
df_sepal.corr(method='pearson').values[1][0]
```
