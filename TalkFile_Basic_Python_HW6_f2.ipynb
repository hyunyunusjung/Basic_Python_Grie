**문제1) 우리는 지난 시간에 setosa의 sepal length에 대해서 하나씩 그래프를 그렸습니다. 다른 iris type의 다른 feature에 대해 똑같이 그래프를 그려볼까요?**

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
feature = iris['data']
labels = iris['target']
feature_names = iris['feature_names']
df = pd.DataFrame(feature, columns=feature_names)
df['target'] = labels
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
ax.boxplot(df0['sepal length (cm)'])
plt.show()
```

**문제2) 우리는 지난 시간에 모든 iris의 sepal length에 대해서 그래프를 겹쳐 그렸습니다. 다른 feature에 대해 똑같이 그래프를 그려볼까요?**

```python
np.arange(len(df0))
plt.figure(figsize=(6,4))
x = np.arange(len(df0))
plt.bar(x, df0['sepal length (cm)'])
plt.title('Sepal length bar graph of iris setosa', fontsize = 14)
plt.show()
```

**문제3) 새로운 와인 데이터를 다뤄봅시다. feature 중 한 가지를 골라서 세 가지의 class를 동시에 그리는 pie chart, box plot, histogram을 그려봅시다. 우리가 그렸던 그래프와 같이 legend도 포함해주세요.**




```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
wine = load_wine()
```

```python
wine
```

```python
arget = wine.target
feature = wine['data']
labels = wine['target']
feature_names = wine['feature_names']
df= pd.DataFrame(feature, columns=feature_names)
df['target'] = labels
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
df['target']=labels
import numpy as np
import matplotlib.pyplot as plt
```

```python
target = wine.target
count_0 = 0
count_1 = 0
count_2 = 0
for i in target:
    if i==0:
        count_0 += 1
    elif i==1:
        count_1 += 1
    else:
        count_2 += 1
plt.figure(figsize = (6,4))
ratio = [count_0, count_1, count_2]
labels = wine.target_names
plt.pie(ratio, labels = labels, autopct='%.1f%%')
plt.legend(loc = 'best', bbox_to_anchor=(1,1))
plt.show()
```

```python
plt.figure(figsize = (6,4))
fig, ax = plt.subplots()
ax.boxplot([df0['alcohol'], df1['alcohol'], df2['alcohol']], labels=labels)
plt.title("Wine dataset Boxplot")
plt.show()
```

```python

plt.figure(figsize = (6,4))
plt.hist(df0['alcohol'], alpha=0.5, label = "class_0")
plt.hist(df1['alcohol'], alpha=0.5, label = "class_0")
plt.hist(df2['alcohol'], alpha=0.5, label = "class_0")
plt.title("Wine dataset Boxplot", fontsize=14)
plt.legend()
plt.show()
```
