**문제5) 정수형 점수 score가 주어질 때, 아래 기준에 따라 학점을 출력해보세요.**

90 이상: A

80 이상 90 미만: B

70 이상 80 미만: C

60 이상 70 미만: D

60 미만: F

```python
#score = 90
if score>=90:
    print("A")
elif (score<90 and score>=80):
    print("B")
elif (score<80 and score>=70):
    print("C")
elif (score<70 and score>=60):
    print("D")
else:
    print("F")
```

**문제6) 정수 n이 주어질 때, 아래 기준에 따라 값을 출력해보세요.**

0이면 "zero"

0이 아니면서 짝수면 "even"

0이 아니면서 홀수면 "odd"

```python
#n=3
if n==0:
    print("zero")
elif (n!=0 and n%2==0):
    print("even")
else:
    print("odd")
```
