## 02 SVM
scikit-learn 와인 데이터셋을 각기 다른 hyperparameter (kernel, C, gamma) 조합의 SVM으로 분류해보고, 가장 성능이 좋았던 hyperparameter 조합을 알려주세요.

https://noanomal.tistory.com/81?
- 위 링크는 와인 데이터셋을 사용하는 방법을 간단하게 정리한 블로그입니다.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
import numpy as np

wine = load_wine()
X = wine.data
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state= 22)

kernels = ['linear', 'poly', 'rbf']
c_value = [1, 5, 10]
gamma_values = [0.1, 1, 10]

b_accuracy = 0
b_capacity = ""

for kernel in kernels:
    for c in c_value:
        for gamma in gamma_values:

            svm = SVC(kernel = kernel, C = c, gamma=gamma)
            svm.fit(X_train, y_train)
            
            accuracy = np.mean(svm.predict(X_test) == y_test)
            
            print(f"kernel={kernel}, C={c}, gamma={gamma}: {accuracy}")
            
            if (accuracy > b_accuracy):
                b_accuracy = accuracy
                b_capacity = f"kernel={kernel}, C={c}, gamma={gamma}"

print(f"최고 성능: {b_capacity}")
print(f"정확도: {b_accuracy}")
```
