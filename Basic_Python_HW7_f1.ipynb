# 01 Titanic dataset preprocessing
1. 필요 없는 column을 지워보세요.
2. 이번엔 age는 mean으로, Embarked는 median으로 imputation해보세요.
3. Sex, Embarked column을 정수 인코딩 해보세요.

```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
del df['PassengerId']
del df['Name']
del df['Ticket']
del df['Cabin']
```

```python
df['Sex']
```

```python
df.columns
```

```python
mapping = {'male':0, 'female':1}
```

```python
df['Sex'] = df['Sex'].map(mapping)
```

```python
df.info()
df.isnull()
import seaborn as sns
df['Age'].mean()
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
embarked_mapping = {'Q':0, 'S':1, 'C':2}
df['Embarked'].map(embarked_mapping)
df['Embarked'] = df['Embarked'].map(embarked_mapping)
```
